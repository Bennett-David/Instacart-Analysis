### Step 1: Import libraries and data

```python
# Import libraries 
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# turning folder path into string
path = r'/Users/davesmac/Desktop/04-2022- Instacart Basket Analysis'
```

```python
# Import orders_products_all df
ords_prods_all = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_all.pkl'))
```

### Step 2: Create bar chart, histogram, scatterplot, and line chart

```python
# Create bar chart
bar = ords_prods_all['orders_day_of_week'].value_counts().sort_index().plot.bar(color =['blue', 'grey', 'blue', 'grey', 'blue', 'grey', 'blue'])
```

```python
# Create histogram
hist = ords_prods_all['prices'].plot.hist(bins = 70)
```

```python
# Create scatterplot 
scatterplot = sns.scatterplot(x = 'prices', y = 'prices',  data = ords_prods_all)
```

### Create line chart with sample of data

```python
#Set seed for random number generator and produce list of random numbers
np.random.seed(4)
dev = np.random.rand(len(ords_prods_all)) <=0.7
```

```python
# Create 'big' sample
big = ords_prods_all[dev]
```

```python
# Create 'small' sample
small = ords_prods_all[~dev]
```

```python
#check subsetting procedure
len(ords_prods_all)
```

```python
len(big) + len(small)
```

#### outputs are equal, meaning sample subsets are correct

```python
# Create df with small sample
df_small = small[['orders_day_of_week', 'prices']]
```

```python
# Create line chart with df_small
line = sns.lineplot(data = df_small, x = 'orders_day_of_week',y = 'prices')
```
