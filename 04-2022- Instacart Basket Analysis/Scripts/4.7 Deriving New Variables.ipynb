### Using user-defined function to create price_label column in subset

```python
# Define a function- create price_label column
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

### Using loc to create column in subset

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range'].value_counts(dropna = False)
```

### Using for-loop to create 'Busiest day' column

```python
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Using for-loop to create new list
result = []
for value in df_ords_prods_merged['orders_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else: 
        result.append('Regular busy')
```

# Step 2: Change 'Busiest day' column to 'Busiest days' 

```python
# Delete 'Busiest day' column
df_ords_prods_merged.drop(columns = ['Busiest_day'])
```

```python
# Use for-loop to create new list
result1 = []
for value in df_ords_prods_merged['orders_day_of_week']:
    if value == 0:
        result1.append('Busiest days')
    elif value == 1: 
        result1.append('Busiest days')
    elif value == 3:
        result1.append('Slowest days')
    elif value == 4:
        result1.append('Slowest days')
    else: 
        result1.append('Regular busy')
```

```python
# Create new column based off of list
df_ords_prods_merged['Busiest_days'] = result1
```

# Step 4: Create 'busiest_period_of_day' column

```python
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Use for-loop to create new list
result2 = []
for value in df_ords_prods_merged['order_hour_of_day']:
    if value == 10:
        result2.append('Most orders')
    elif value == 11:
        result2.append('Most orders')
    elif value == 2: 
        result2.append('Fewest orders')
    elif value == 3: 
        result2.append('Fewest orders')
    elif value == 4: 
        result2.append('Fewest orders')
    elif value == 5: 
        result2.append('Fewest orders')
    else: 
        result2.append('Average orders')
```

#### *Note- I determined hours 10 & 11 to be the busiest, while hours 2, 3, 4, & 5 to be least busiest.  

```python
# Create new columb based off of list
df_ords_prods_merged['busiest_period_of_day'] = result2
```
